
# Problem

* An orthopaedist is a medical doctor specializing in diagnosing and treating disorders related to the skeletal system. Part of their job is to distinguish between a healthy person and a person with Osteoarthritis by looking at their knee X-ray images.

* Osteoarthritis: Osteoarthritis, commonly known as wear-and-tear arthritis, is a condition in which the natural cushioning between joints — cartilage — wears away. When this happens, the bones of the joints rub more closely against one another with less of the shock-absorbing benefits of cartilage. The rubbing results in pain, swelling, stiffness, decreased ability to move, and, sometimes, the formation of bone spurs.

# Task

Your task is to create a deep learning model that can provide the probability of having osteoarthritis for a given knee X-ray image.


```python
#Import libraby

import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import models
from tensorflow.keras import optimizers
from keras.layers import Conv2D, MaxPool2D, Flatten, Dense
```

```python
# Import Data
train_dir = r'C:/Users/Shankar/Desktop/ML/Intern/assignment_2/Osteoarthritis_Assignment_dataset/train'
test_dir = r'C:/Users/Shankar/Desktop/ML/Intern/assignment_2/Osteoarthritis_Assignment_dataset/test'
val_dir = r'C:/Users/Shankar/Desktop/ML/Intern/assignment_2/Osteoarthritis_Assignment_dataset/Valid'
```

# Data Argumentation Method

IN this method limited data is to apply different transformations on the available data to synthesize new data. Approach of synthesizing new data from the available data is knows as  ‘Data Augmentation’.

```python
# Generating batches of tensor image data

train_gen = ImageDataGenerator(rescale=1./255,
                               zoom_range=.25,
                               shear_range=.25,
                               horizontal_flip=True,
                                rotation_range=40,)
val_gen = ImageDataGenerator(rescale=1./255)

train_set = train_gen.flow_from_directory(
    train_dir,
    target_size=(150, 150),
    batch_size=20,
    class_mode='binary')

val_set = val_gen.flow_from_directory(
    val_dir,
    target_size =(150, 150),
    batch_size = 20,
    class_mode = 'binary')
```

# CNN Model for predict osteoarthritis for a given knee X-ray image

### CNN architecture

```python
# Generate CNN layers and using MAXPooling
cl = models.Sequential()

cl.add(Conv2D(32, (3, 3) , input_shape=(150, 150, 3), activation='relu')) 

# pooling layer
cl.add(MaxPool2D((2,2)))



cl.add(Conv2D(128, (3,3), activation='relu'))

cl.add(MaxPool2D((2,2)))


# Flatting
cl.add(Flatten())

# full connection 
cl.add(Dense(units=256, activation='relu'))


#for output(binary classification)
cl.add(Dense(units=1, activation='sigmoid'))



# compiling model

cl.compile(optimizer=optimizers.RMSprop(learning_rate=1e-5), loss = 'binary_crossentropy', metrics=['acc'])

# use a early stop
early_stopping = tf.keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True)

# Model Training

model = cl.fit(train_set,
                steps_per_epoch=118,
                epochs = 20,
                validation_data= val_set,
                validation_steps=32,
                callbacks= [early_stopping])

#Steps_per_epoch = n_samples / batch_size

```

```python
model_history = pd.DataFrame(model.history)
model_history[:2]
```

```python
model_history.plot(figsize = (8,5))
plt.grid(True)
plt.gca().set_ylim(0,1);
```

# Model Testing

```python
test_gen = val_gen.flow_from_directory(test_dir,
                                       target_size = (150, 150),
                                       batch_size = 20,
                                       class_mode ="binary")
cl.evaluate(test_gen, steps=42)
```

```python
import tensorflow.keras.preprocessing.image
from IPython.display import Image

def image_prediction(image_path, model):
    test_image = keras.utils.load_img(image_path, target_size = (150, 150))
    image = Image(filename=image_path)
    test_image = tf.keras.utils.img_to_array(test_image)
    test_image = np.expand_dims(test_image, axis = 0)
    result = model.predict(test_image)
    
    print(result)
        
    return image
```

#   VGC16 Model for predict osteoarthritis for a given knee X-ray image

```python
from tensorflow.keras.applications import VGG16

VGG16_base = VGG16(weights = 'imagenet',
                    include_top = False,
                    input_shape = (150, 150, 3))
```

```python
VGG16_base.summary()
```

```python
#VGG model layes and flattening
vgg_model = models.Sequential()
vgg_model.add(VGG16_base)
vgg_model.add(Flatten())
vgg_model.add(Dense(512,activation='relu'))
vgg_model.add(Dense(1, activation='sigmoid'))
```

```python
vgg_model.summary()
```

```python
# Model Compile
vgg_model.compile(loss = 'binary_crossentropy',
             optimizer = optimizers.RMSprop(learning_rate=1e-4),
             metrics=['acc'])
```

```python
# Model Training
early_stop = tf.keras.callbacks.EarlyStopping(patience=4,restore_best_weights=True)
model1 = vgg_model.fit(train_set,
                       steps_per_epoch = 118,
                       epochs = 30,
                       validation_data = val_set,
                       callbacks= [early_stop])
```

```python
pd.DataFrame(model1.history).plot(figsize = (8,5))
plt.grid(True)
plt.gca().set_ylim(0,1)
plt.show()
```

# Model Testing

```python
vgg_test = val_gen.flow_from_directory(test_dir,
                                       target_size = (150, 150),
                                       batch_size = 20,
                                       class_mode ="binary")
vgg_model.evaluate(vgg_test, steps = 42)
```
